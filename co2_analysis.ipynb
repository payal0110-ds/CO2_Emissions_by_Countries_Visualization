```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
import os
```

```python
dir_name=os.getcwd()
dir_name
```

```python
for files in os.listdir(dir_name):
    if files.endswith('.csv'):
        file_path=os.path.join(dir_name,files)
print(file_path)
```

```python
df=pd.read_csv(file_path,index_col='Country Name')
df.head()
```

```python
df.info()
```

```python
df1=df.transpose()
df1.index=pd.to_datetime(df1.index).year
df1.head()
```

```python
fig_1=px.line(data_frame=df1, x=df1.index, y=['World'], markers=True)
fig_1
```

```python
with open('line_chart.pkl','wb') as file:
    pickle.dump(fig_1,file)
```

```python
df.head()
```

```python
df.reset_index(level=0,inplace=True)
df.head()
```

```python
df2=df.melt(id_vars=['Country Name'], var_name='Year').sort_values(by=['Year'])
df2.head()
```

```python
fig_2=px.choropleth(data_frame=df2[df2['Country Name']!='World'],locationmode='country names',
                    locations='Country Name', color='value', animation_frame='Year',
                    color_continuous_scale=px.colors.sequential.RdBu_r,range_color=(200,0))
fig_2
```

```python
with open('map_graph.pkl','wb') as file:
    pickle.dump(fig_2,file)
```

```python
from raceplotly.plots import barplot
race_plot=barplot(df=df2[df2['Country Name']!='World'],item_column='Country Name', value_column='value',time_column='Year', top_entries=10)
fig_3=race_plot.plot(title='Carbon Emission Race by Country (Since 1960 to 2018)', value_label='Value',item_label='Top 10 Countries')
fig_3
```

```python
with open('bar_plot.pkl','wb') as file:
    pickle.dump(fig_3,file)
```

```python
country_col=df2[df2['Country Name']!='World'].groupby(by=['Country Name']).max().sort_values(by=['value'], ascending=False).head(10).index.to_list()
country_col.extend(['India','China'])
sample_mean=df2[df2['Country Name'].isin(country_col)]['value'].mean()
pop_mean = df2['value'].mean()
```

```python
fig_4, ax = plt.subplots(figsize=(15, 8))
sns.kdeplot(data=df2[df2['Country Name'].isin(country_col)],x='value',hue='Country Name',fill=True,linewidth=3,edgecolor='white',alpha=0.7,ax=ax)
ax.axvline(x=sample_mean)
ax.axvline(x=pop_mean)
ax.text(x=sample_mean, y=0.015, s='Sample mean')
ax.text(x=pop_mean, y=0.015, s='Population mean')
ax.set_xlim(-20, 60)
ax.set_ylim(0.00, 0.04)
```

```python
with open('kde_plot.pkl','wb') as file:
    pickle.dump(fig_4,file)
```

```python
fig_5=px.box(data_frame=df2[df2['Country Name'].isin(country_col)], x='value', y='Country Name', color='Country Name')
fig_5
```

```python
with open('box_plot.pkl','wb') as file:
    pickle.dump(fig_5,file)
```

```python
df2['decade']=1
df2['decade'][df2['Year'].between('1960','1970')]='1960-1970'
df2['decade'][df2['Year'].between('1971','1980')]='1971-1980'
df2['decade'][df2['Year'].between('1981','1990')]='1981-1990'
df2['decade'][df2['Year'].between('1991','2000')]='1991-2000'
df2['decade'][df2['Year'].between('2001','2010')]='2001-2010'
df2['decade'][df2['Year'].between('2011','2020')]='2011-2018'
```

```python
fig_6=px.box(data_frame=df2[df2['Country Name'].isin(country_col)], x='value', y='Country Name', color='Country Name', facet_col='decade')
fig_6
```

```python
with open('box_plot_decade.pkl','wb') as file:
    pickle.dump(fig_6,file)
```

#### As Aruba contributed high carbon emssion from 1960 - 1990, shown in this chart, after 1990, there is not much carbon emission from this country, hence let us remove 'Aruba' to clearly view other countries details.

```python
country_col.remove('Aruba')
```

```python
fig_7=px.box(data_frame=df2[df2['Country Name'].isin(country_col)], x='value', y='Country Name', color='Country Name', title='Variation of Carbon emission by top 11 countries', facet_col='decade')
fig_7
```

```python
fig_8=px.bar(data_frame=df2[df2['Country Name'].isin(country_col)].sort_values(by=['value'],ascending=False), x='value',y='Country Name', facet_col='decade', color='Country Name',barmode='stack' )
fig_8
```

```python
with open('box_bar.pkl','wb') as file:
    pickle.dump(fig_8, file)
```
